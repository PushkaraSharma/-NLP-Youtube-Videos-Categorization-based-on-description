```python
import string
from nltk.corpus import stopwords
from nltk import word_tokenize
stop_words = stopwords.words('english')
def bag_words(words):
    words = words.encode('ascii','ignore')
    words = str(words)
    words = words.lower()
    processed = words.replace(r'^\w+@[a-zA-Z_]+?\.[a-zA-Z]{2,3}$', 'email')
    processed = processed.replace(r'^\D?(\d{3})\D?\D?(\d{3})\D?(\d{4})$', 'phoneno')
    processed = processed.replace(
        r'^(http(s?)\:\/\/)*[0-9a-zA-Z]([-.\w]*[0-9a-zA-Z])*(:(0-9)*)*(\/?)([a-zA-Z0-9\-\.\?\,\'\/\\\+&amp;%\$#_]*)?$',
       'website')
    processed = processed.replace(r'$|\£|\€', 'money')
    processed = processed.replace(r'\d+(\.\d+)?', 'number')
    processed = processed.replace(r'^\s+|\s+?$', '')
    processed = processed.replace(r'\s+', ' ')
    processed = processed.replace(string.punctuation,'')
    processed = processed.replace('=','')
    processed = processed.replace('|','')
    processed = processed.replace('.','')
    processed = processed.replace('!','')
    tokens = word_tokenize(words)
    clean = []
    for i in tokens:
        if i not in stop_words and i not in string.punctuation and i.startswith('//')==False:
            clean.append(i)
    dicton = dict([word, True] for word in clean)
    return dicton
        
```

```python
string.punctuation
```

```python
final = []
for i in range(len(X)):
    final.append((bag_words(X[i]),Y[i]))
```

```python
final
```

```python
from sklearn.model_selection import train_test_split
training,testing = train_test_split(final,test_size = 0.20)
```

```python
from nltk import NaiveBayesClassifier
from nltk import classify
```

```python
classifier = NaiveBayesClassifier.train(training)
acc = classify.accuracy(classifier,testing)
acc
```
